## Explainatory Data Analysis (EDA)

### Exploratory data analysis (EDA) is used by data scientists to analyze and investigate data sets and summarize their main characteristics, often employing data visualization methods.

### The main purpose of EDA is to help look at data before making any assumptions. It can help identify obvious errors, as well as better understand patterns within the data, detect outliers or anomalous events, find interesting relations among the variables.

## Types:

### 1. Univariate Analysis: This focuses on analyzing one variable at a time. It involves exploring the characteristics of only one single variable.
### 2. Bivariate Analysis: Bivariate analysis involves exploring the relationship between two variables. It involves exploring the characteristics of two variables at a time.
### 3. Multivariate Analysis: Multivariate analysis examines the relationships between multiple variables simultaneously.

# 1. Categorical Data

## a. Countplot

```python
# Counting the number of people boarded from each port
train['Embarked'].value_counts()
```

### As we can see there 644 people embarked from Southampton, 168 people embarked from Cherbourg and 77 people boarded from Queenstown

```python
# Plotting the
sns.countplot(train['Embarked'], legend='auto')
```

### As we can see a clear picture of countplot displaying the number of people boarded from each port

## b. PieChart

```python
# Pie chart to show percentage of people comes under male or female
train['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

### Here it shows 64.76% of the population travelling were males and 35.24% were females.

# 2. Numerical Data

## a. Histogram:
###  A histogram is a graph showing frequency distributions. It is a graph showing the number of observations within each given interval. 
### Each interval is called a bin in X-axis and the Y-axis of the histogram will show the frequency of data points in each bin.
### Remember we are doing univariate analysis now so all these operations will be performed in a single column/variable of the data.

```python
import matplotlib.pyplot as plt
plt.hist(train['Age'],bins=5)
```

### I know reading and taking insights from histogram is a bit tricky at start but don't worry we'll pull it through.
### In the above histogram in X-axis it is the age of people commuted in titanic but in Y-axis it is the number of people belong to the age group of X-axis.
### We can also see that the histogram can be divided into 5-bins/5-intervals like 10-18 has 1bin, 18-32 has 1bin etc.
### You can see around 350 of the people travelling in the ship belong to the age group of 18-32. So this is how you read one histogram of a numerical variable.

## b. Distplot

### It is used basically for univariant set of observations and visualizes it through a histogram i.e. only one observation and hence we choose one particular column of the dataset

```python
sns.distplot(train['Age'])
```

### We can think of distplot like another histogram but instead of showing numbers in the Y-axis it shows probability distribution of the particular column in the Y-axis.
### We can see more than 30% of the people is in the age group 22-25

## c. Boxplot

### Boxplot is basically a 5 point summary plot can be used to fetch insights from a numerical variable/column.
### It gives the following values (Minimum, Maximum, 25th Percentile, 75th Percentile and the Interquartile Range IQR(75th - 25th)percentile).
### It also gives the vision of how outliers are placed in the variable/column. 

```python
# Displaying Box Plot
sns.boxplot(train['Age'])
```

### In the above boxplot we can see the median, min, max, q1, q3, IQR and possible outliers for the column/variable age.

```python
# validating the minimum age
train['Age'].min()
```

```python
# validating the maximum age
train['Age'].max()
```

```python
# validating the median age
train['Age'].median()
```

### We can verify from the above numbers the boxplot is intact with the correct frequency.
